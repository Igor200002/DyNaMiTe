```
import random
import timeit
def random_arr(length=100):  # генерирует рандомный массив массив
    arr = []
    while len(arr) < length:
        arr.append(random.randint(-1000, 1000))
    return arr
init_arr = random_arr(1_000)  # Генерация массива рандомного массива из 1000 элементов
print(init_arr) # выводим для наглядности
def default_sort(arr):  # дефолтная сортировка через sort
    arr.sort()
    return arr

def vs_sort(arr):  # Сортировка вставками
    for i in range(len(arr)):  # Идём вперёд с начала массива
        if i != len(arr) - 1 and arr[i] > arr[i + 1]:
            arr[i], arr[i + 1] = arr[i + 1], arr[i] # если нынешний элемент больше следующего меняем их местами
            for j in range(len(arr) - 1, 0, -1):  # Идём назад с конца массива
                if arr[j] <= arr[j - 1]:
                    arr[j - 1], arr[j] = arr[j], arr[j - 1] # если прошлый элемент нынешнего меняем их местами
    return arr
# Сортировка массива и вывод времени
print(vs_sort(init_arr))
print("Сортировка по умолчанию   " + str(timeit.timeit("default_sort(init_arr)", number=1, globals=globals())))
print("Сортировка вставками      " + str(timeit.timeit("vs_sort(init_arr)", number=1, globals=globals())))
```
